# Fast Style Transfer

> ***Style transfer allows you to take famous paintings, and recreate your own images in their styles!***


Neural Style Transfer algorithms make use of deep neural networks to perform the image transformation, such as creating artificial artwork from photographs, by transferring the style of famous paintings to photographs.

This fun project from the Udacity Deep Learning Nanodegree uses the fast-style-transfer GitHub repo by Logan Lengstrom,  Either use `git` to clone the repository, or you can download the whole thing as a Zip archive and extract it.

I am going to use it to style the photograph of the lovely ちぃちゃん in the style of different artists. 



To access the files easily from one place, I first forked the original repo, then added my  image files and the two checkpoint files, Rain Princess and Wave to that repo. 


The checkpoints were trained on the following paintings:


*  Rain Princesss, by Leonid Afremov
* The Great Wave off Kanagawa, by Hokusai
